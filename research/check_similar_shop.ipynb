```python
text_sim = pd.DataFrame(similarity_matrix, columns=cluster_sample.index, index=cluster_sample.index)
text_sim.head()
```

```python
similar_text_indexes = {}
for idx in text_sim.index:
    sorted_similarities = text_sim.loc[idx].sort_values(ascending=False)
    filtered_similarities = sorted_similarities[(sorted_similarities > 0.5) & (sorted_similarities < 1)]
    
    similar_text_indexes[idx] = [filtered_similarities]

similar_text_indexes
```

```python
cluster_sample['same_shop_group'] = pd.NA

def calculate_distance(lat1, lon1, lat2, lon2):
    return geodesic((lat1, lon1), (lat2, lon2)).meters

threshold = 0.95
distance_threshold = 100

group_counter = 1  
assign_shops = set()
cluster_sample.reset_index(inplace=True)

for i in range(len(similarity_matrix)):
    if pd.isna(cluster_sample.iloc[i]['same_shop_group']) and cluster_sample.index[i] not in assign_shops:
        cluster_sample.loc[cluster_sample.index[i], 'same_shop_group'] = group_counter

        assign_shops.add(cluster_sample.index[i])

        for j in range(len(similarity_matrix)):
            if i != j and similarity_matrix[i][j] > threshold:
                lat1, lon1 = cluster_sample.iloc[i][['Latitude', 'Longitude']]
                lat2, lon2 = cluster_sample.iloc[j][['Latitude', 'Longitude']]
                distance = calculate_distance(lat1, lon1, lat2, lon2)
                
                if distance < distance_threshold:
                    if pd.isna(cluster_sample.iloc[j]['same_shop_group']):
                        cluster_sample.loc[cluster_sample.index[j], 'same_shop_group'] = group_counter
                        assign_shops.add(cluster_sample.index[j])

        group_counter += 1

cluster_sample.head()
```

```python
similarity_matrix = np.ones((len(sentences),len(sentences)))
nlp = spacy.load("en_core_web_md")
```

```python
for i in range(len(sentences)):
    shop_1 = sentences[i]
    for j in range(i,len(sentences)):
        shop_2 = sentences[j]
        if i!=j:
            doc1 = nlp(shop_1)
            doc2 = nlp(shop_2)
            similarity = doc1.similarity(doc2)
            similarity_matrix[i][j]= similarity
            similarity_matrix[j][i]= similarity
```
