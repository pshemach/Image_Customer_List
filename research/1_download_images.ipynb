```python
import pandas as pd
import requests
from PIL import Image
from io import BytesIO
from os import walk
import threading
import time
```

```python
df = pd.read_csv('../data/data.csv')
```

```python
data = df.iloc[:100]
data.head()
```

```python
image_link_lis = data['image'].to_list()
```

```python
root_image_link = 'https://surgepilot.lk/'
```

```python
def get_image_link(lis):
    image_link = []
    root_image_link = 'https://surgepilot.lk/'
    for img in lis:
        image_name = img.split('/')[-1]
        image_link.append(root_image_link + image_name)
    return image_link
```

```python
updated_img_lis = get_image_link(image_link_lis)
```

```python
image_save_dir = '../data/images'
```

```python
def download_images(save_dir,dir_lis, img_lis):
    for i, img in enumerate(img_lis):
        img_name = img.split("/")[-1]
        if img_name not in dir_lis:
            response = requests.get(img)
            print(f'---Downloading--- Image: {image_name}')
            img = Image.open(BytesIO(response.content))
            img.save(f"{save_dir}/{img_name}")
```

```python
img_in_dir = []
for dirpath, dirnames, filenames in walk(image_save_dir):
    img_in_dir.extend(filenames)
    

start = time.time()

# download_images(image_save_dir, img_in_dir,updated_img_lis[:40])
thread = list()

for i in range(4):
    x = threading.Thread(
        target=download_images, args=(image_save_dir, img_in_dir, updated_img_lis,)
    )
    thread.append(x)
    x.start()
    
for t in thread:
    t.join()
    
end = time.time()

print(f"---Total time taken to download images: {end - start} seconds ---")
```
